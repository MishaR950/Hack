```python
from sklearn.linear_model import LogisticRegressionCV
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.preprocessing import StandardScaler
import numpy as np

# Предположим, что df уже определён и содержит необходимые данные
df = pd.get_dummies(df)  # преобразовываем все категориальные переменные
scaler = StandardScaler()

# Теперь мы предполагаем, что все переменные в df, кроме 'credit_score', являются числовыми.
X = df.drop('credit_score', axis=1)
y = df['credit_score']

# Масштабируем только числовые переменные.
scaled_features = scaler.fit_transform(X.select_dtypes(include=[np.number]))

scaled_data = pd.DataFrame(scaled_features, index=X.index, columns=X.select_dtypes(include=[np.number]).columns)
# Не числовые данные нам не надо масштабировать, но они уже one-hot encoded, присоединим их к scaled_data
scaled_data = scaled_data.join(X.select_dtypes(exclude=[np.number]))

# Далее ваш код для разделения на обучающую и тестовую выборки и так далее.
scaler = StandardScaler()
scaler.fit(df.drop('credit_score', axis=1))
scaled_features = scaler.transform(df.drop('credit_score', axis=1))
scaled_data = pd.DataFrame(scaled_features, columns=df.drop('credit_score', axis=1).columns)

x = scaled_data
y = df['credit_score']

# Делим на обучающую и тестовую выборки
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.3, random_state=0)

# Тренируем модель
lr = LogisticRegressionCV(cv=3, random_state=0)
lr.fit(x_train, y_train)

from sklearn.metrics import precision_score, recall_score

y_pred_lr = lr.predict(x_test)

# Убедитесь, что ваши данные подходят под задачу классификации перед вычислением точности и полноты
precision_lr = precision_score(y_test, y_pred_lr, average='binary') # Для многоклассовой добавьте параметр average
recall_lr = recall_score(y_test, y_pred_lr, average='binary')       # Например, average='micro' или average='macro'

# Выводим результаты и видим, что модель работает нормально, но не идеально
print('LogisticRegressionCV: точность = {:.2f}, полнота = {:.2f}'.format(precision_lr, recall_lr))

print("Правильность на обучающем наборе: {:.2f}".format(lr.score(x_train, y_train)))
print("Правильность на тестовом наборе: {:.2f}".format(lr.score(x_test, y_test)))
```

```python
from sklearn.ensemble import RandomForestClassifier
import pandas as pd
import numpy as np

df = pd.read_csv('/content/credit_score_df.csv', sep = '|')
X = df.iloc[:, :-1].values

from sklearn.preprocessing import OneHotEncoder

# Предположим, что df - это ваш исходный DataFrame, и 'occupation' - это колонка для кодирования.
df_onehot = df.copy()

# Создаем экземпляр OneHotEncoder с параметром drop='first'
onehotencoder = OneHotEncoder(drop='first', sparse=False)

# Проводим one-hot кодирование, исключая первый признак, и создаем новый DataFrame
encoded_occupation = pd.DataFrame(onehotencoder.fit_transform(df_onehot[['occupation']]))

# Получаем новые имена для one-hot закодированных колонок, исключая первый признак
encoded_occupation.columns = onehotencoder.get_feature_names_out(['occupation'])

# Присоединяем закодированный DataFrame к исходному
df_onehot = df_onehot.join(encoded_occupation)

# Удаляем исходную колонку 'occupation'
df_onehot.drop('occupation', axis=1, inplace=True)

# Теперь df_onehot содержит закодированные признаки вместо исходной колонки 'occupation'
df_onehot

# Создаем массив данных
X = ['month',	'age',	'occupation',	'annual_income',	'monthly_inhand_salary',	'num_bank_accounts',	'num_credit_card',	'num_of_loan',	'num_credit_inquiries',	'credit_history_age',	'amount_invested_monthly',	'payment_behaviour',	'monthly_balance']
y_train = ['credit_score']
# Преобразуем массив данных в одномерный массив
X_flattened = np.ravel(X)

# Создаем объект RandomForestClassifier
clf = RandomForestClassifier(n_estimators=100)

# Обучаем модель на данных
clf.fit(X_flattened, y_train)

# Получаем значения важности признаков
feature_imp = pd.Series(clf.feature_importances_, X_flattened).sort_values(ascending=False)

# Используем iloc для доступа к отдельным значениям массива весов и создания Series на основе этих значений
feature_imp_reshaped = feature_imp.iloc[:, :].reshape(-1, 1)

# Выводим значения важности признаков
print(feature_imp_reshaped)

```
